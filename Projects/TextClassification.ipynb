# 3) Clean & Transform Data

```python
import re
import nltk
# Uncomment to download "stopwords"
nltk.download("stopwords")
from nltk.corpus import stopwords

def text_preprocessing(s):
    """
    - Lowercase the sentence
    - Change "'t" to "not"
    - Remove "@name"
    - Isolate and remove punctuations except "?"
    - Remove other special characters
    - Remove stop words except "not" and "can"
    - Remove trailing whitespace
    """
    s = s.lower()
    # Change 't to 'not'
    s = re.sub(r"\'t", " not", s)
    # Remove @name
    s = re.sub(r'(@.*?)[\s]', ' ', s)
    # Isolate and remove punctuations except '?'
    s = re.sub(r'([\'\"\.\(\)\!\?\\\/\,])', r' \1 ', s)
    s = re.sub(r'[^\w\s\?]', ' ', s)
    # Remove some special characters
    s = re.sub(r'([\;\:\|•«\n])', ' ', s)
    # Remove stopwords except 'not' and 'can'
    s = " ".join([word for word in s.split()
                  if word not in stopwords.words('english')
                  or word in ['not', 'can']])
    # Remove trailing whitespace
    s = re.sub(r'\s+', ' ', s).strip()

    return s
```

## Clean and Transform Training Data

```python
data = []
for i in range(0, df_train.shape[0]):
  text_and_class = []
  title = df_train.iloc[i][0]
  abstract = df_train.iloc[i][1]
  classes = df_train.iloc[i][2]

  title_abstract = title + " " + abstract # concat title and abstract

  text_and_class.append(text_preprocessing(title_abstract))
  text_and_class.append(classes)

  data.append(text_and_class)
```

```python
new_df_train = pd.DataFrame(data = data, columns = ["Text", "Classes"])
```

```python
new_df_train.head()
```

## Clean and Transform Testing Data

```python
data2 = []
for i in range(0, df_test.shape[0]):
  text = []
  title = df_test.iloc[i][0]
  abstract = df_test.iloc[i][1]

  title_abstract = title + " " + abstract # concat title and abstract

  text.append(text_preprocessing(title_abstract))

  data2.append(text)
```

```python
new_df_test = pd.DataFrame(data = data2, columns = ["Text"])
```

```python
new_df_test.head()
```

# 4) Encoding the classes in new_df_train

```python
from sklearn.preprocessing import MultiLabelBinarizer

categories = ["CE","ENV","BME","PE","METAL","ME","EE","CPE","OPTIC",
              "NANO","CHE","MATENG","AGRI","EDU","IE","SAFETY","MATH","MATSCI"]

mlb = MultiLabelBinarizer(classes=categories)

# Transform the classes column into binary indicators
classes_encoded = mlb.fit_transform(new_df_train['Classes'])

# Create endcoded dataframe
encoded_df = pd.DataFrame(classes_encoded, columns=categories)

# Concatenate the new_df_train and encoded_df together
df_encoded = pd.concat([new_df_train.drop(columns=['Classes']), encoded_df], axis=1)
```

# 5) Split train/validation with 20%, seed = 42

```python
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(df_encoded['Text'], df_encoded.drop('Text', axis=1), test_size=0.2, random_state=42)
```

# 6) Feature Extraction with TFIDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Implement TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_valid_tfidf = tfidf_vectorizer.transform(X_valid)
```

# 7) Train Naive Bayes Classifier with Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.multioutput import MultiOutputClassifier

# Train Naive Bayes Classifier with Hyperparameter Tuning for Multilabel Classification
param_grid = {'estimator__alpha': [0.1, 0.5, 1.0, 2.0]}
grid_search = GridSearchCV(estimator=MultiOutputClassifier(MultinomialNB()), param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train_tfidf, y_train)
```

```python
# Get the best hyperparameters
best_alpha = grid_search.best_params_['estimator__alpha']
print(best_alpha)
```

```python
# Train the Naive Bayes classifier with the best hyperparameters
nb_classifier_tuned = MultiOutputClassifier(MultinomialNB(alpha=best_alpha))
nb_classifier_tuned.fit(X_train_tfidf, y_train)
```

# 8) Model Weight & Save the model

```python
model_weights = nb_classifier_tuned.estimators_
```

```python
model_weights
```

```python
import joblib

joblib.dump(nb_classifier_tuned, 'model.pkl')
```

# 9) Predictions on the validation set and evaluate

```python
from sklearn.metrics import classification_report

# Predictions on the validation set and evaluate
y_pred_valid = nb_classifier_tuned.predict(X_valid_tfidf)
print("Validation Set Classification Report:")
print(classification_report(y_valid, y_pred_valid))
```
