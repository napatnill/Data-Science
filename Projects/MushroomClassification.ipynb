```python
nominal_variables = df.dtypes[df.dtypes == 'object'].index.tolist()
```

```python
nominal_variables
```

```python
# 1.) get dummy dataframe
dummy_dataframes = []
drop_nominal_cols = []
for col in nominal_variables:
  drop_nominal_cols.append(col)
  dummy_dataframe = pd.get_dummies(df[col], drop_first=True)
  dummy_dataframes.append(dummy_dataframe)
```

```python
# 2.) drop original nominal variable
df.drop( drop_nominal_cols, axis=1, inplace=True)
```

```python
# 3.) concat dummy dataframe to original dataframe
concat_dataframe_list = [df] + dummy_dataframes # merge two lists together
df_with_dummy = pd.concat( concat_dataframe_list , axis=1)
```

```python
df_with_dummy.head()
```

# 8) Split train/test with 20% test, stratify, and seed = 2020.

```python
from sklearn.model_selection import train_test_split

y = df_with_dummy['label'] # or can use df_with_dummy.pop('label')
X = df_with_dummy.drop('label', axis=1) # don't change the original value of df_with_dummy
```

```python
y
```

```python
X.head()
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.20, random_state=2020)
```

```python
print(f'X_train : {X_train.shape}')
print(f'y_train : {y_train.shape}')
print(f'X_test : {X_test.shape}')
print(f'y_test : {y_test.shape}')
```

```python
# [From step 8], how much is each training and testing sets
print(f'({y_train.shape[0]}, {y_test.shape[0]})')
```

# 9) Create a Random Forest with GridSearch on training data with 5 CV.
<u>note</u>: 5 CV = 5-fold cross validation
- 'criterion' : ['gini','entropy']
- 'max_depth' : [2,3,6]
- 'min_samples_leaf' : [2,5,10]
- 'n_estimators' : [100,200]
- 'random_state' : 2020

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier(random_state=2020)
```

```python
# make a dictionary of hyperparameter to search
search_space = {
    'n_estimators' : [100, 200],
    'criterion' : ['gini', 'entropy'],
    'max_depth' : [2, 3, 6],
    'min_samples_leaf' : [2, 5, 10]
}
```

```python
# make a GridSearchCV object
from sklearn.model_selection import GridSearchCV

GS = GridSearchCV(
    estimator=RF_model,
    param_grid=search_space,
    cv=5
)
```

```python
# train model
GS.fit(X_train, y_train)
```

```python
print(GS.best_estimator_) # to get the complete details of the best model
```

```python
GS.best_params_ # to get the only best hyperparameter
```

```python
# [From step 9] best params after doing random forest grid search
GS.best_params_
```

# 10) Predict the testing data set with confusion_matrix and classification_report.

```python
# Use the best hyperparameters to create the final Random Forest model
best_RF_model = RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    max_depth=6,
    min_samples_leaf=2,
    random_state=2020
)
```

```python
# Train the final model on the entire training dataset
best_RF_model.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = best_RF_model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, classification_report

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
# Generate and print the classification report
report = classification_report(y_test, y_pred)
print("Classification Report:")
print(report)
```
