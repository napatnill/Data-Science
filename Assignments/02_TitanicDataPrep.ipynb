# 4) Handle outliers

For the variable Fare, replace outlier values with the boundary values


*   If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
*   If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)

<font color='blue'>Q4) What is the mean of Fare after replacing the outliers (round 2 decimal points)?</font>



```python
df['Fare'].head()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR

print(f'Q1 = {Q1}\nQ3 = {Q3}\nIQR = {IQR}\n')
print(f'Q1 - 1.5IQR = {lower_bound}')
print(f'Q3 + 1.5IQR = {upper_bound}')
```

```python
df['Fare']
```

```python
# remove outliers

# If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
df['Fare'] = df['Fare'].apply(lambda x : lower_bound if x < lower_bound else x )

# If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)
df['Fare'] = df['Fare'].apply(lambda x : upper_bound if x > upper_bound else x )
```
