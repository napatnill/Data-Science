## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
# use code pattern from "4. Advanced Pandas (Data Set:Trending YouTube Video Statistics)"
cat_df = pd.read_json('data/US_category_id.json')
```

```python
import json

with open('data/US_category_id.json') as fd:
    cat = json.load(fd)

cat_list = []
for d in cat['items']:
    cat_list.append((int(d['id']), d['snippet']['title']))
```

```python
cat_df = pd.DataFrame(cat_list, columns=['id', 'category'])
```

```python
cat_df
```

```python
vdo_df_withcat = vdo_df.merge(cat_df, left_on='category_id', right_on='id')
```

```python
vdo_df_withcat.head(3)
```

```python
vdo_df_withcat[['title', 'category_id', 'category']]
```

```python
vdo_df_withcat[ vdo_df_withcat['category'] == 'Sports' ].head(3)
```

```python
# Note:
# category "Sports" -> category_id = 17
# category "Comedy" -> category_id = 23
```

```python
sports_df = vdo_df_withcat[ vdo_df_withcat['category'] == 'Sports' ]
comedy_df = vdo_df_withcat[ vdo_df_withcat['category'] == 'Comedy' ]
```

```python
# focus on trending_date
total_views_sports = sports_df.groupby('trending_date')['views'].sum()
total_views_comedy = comedy_df.groupby('trending_date')['views'].sum()
```
