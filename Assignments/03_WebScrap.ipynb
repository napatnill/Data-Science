```python
# create function to receive website url to get list of holy days
# (because the 3 websites have the same format)

def get_holy_days(url):
  web = requests.get(url)
  soup = BeautifulSoup(web.text, "lxml")
  data = []
  for line in soup.find_all('div', attrs={'class': 'bud-day-col'}):
    text = line.get_text()
    if ( ( 'วัน' in text ) and ( '(' not in text ) ):
      data.append(text)
  return data
```

```python
holy_days_65 = get_holy_days('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2565.aspx')
holy_days_66 = get_holy_days('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2566.aspx')
holy_days_67 = get_holy_days('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2567.aspx')
```

```python
dict_holy_days_65 = { 'holy_day' : holy_days_65 }
dict_holy_days_66 = { 'holy_day' : holy_days_66 }
dict_holy_days_67 = { 'holy_day' : holy_days_67 }
```

```python
# Example to convert day to datetime
# weekday() returns day of week with 0=Monday, ..., 6=Sunday.

import dateparser

dt = dateparser.parse('วันเสาร์ที่ 21 กันยายน 2564')
dt = dt.replace(year=dt.year-543)
print(dt)
print(dt.weekday())
```

```python
# create function to add date_time column to dataframe

import dateparser

def add_datetime(df):
  date_time = []
  for holy_day in df['holy_day']:
    dt = dateparser.parse(holy_day)
    dt = dt.replace(year=dt.year-543)
    date_time.append(dt)
  df['date_time'] = date_time
```

```python
# create function to add day column to dataframe

def day(weekday):
  if weekday == 0:
    return 'Monday'
  if weekday == 1:
    return 'Tuesday'
  if weekday == 2:
    return 'Wednesday'
  if weekday == 3:
    return 'Thursday'
  if weekday == 4:
    return 'Friday'
  if weekday == 5:
    return 'Saturday'
  if weekday == 6:
    return 'Sunday'
  return 'Not in Day range'

def add_day(df):
  days = []
  for dt in df['date_time']:
    days.append(day(dt.weekday()))
  df['day'] = days
```

```python
add_datetime(df_holy_days_65)
add_datetime(df_holy_days_66)
add_datetime(df_holy_days_67)
```

```python
add_day(df_holy_days_65)
add_day(df_holy_days_66)
add_day(df_holy_days_67)
```
